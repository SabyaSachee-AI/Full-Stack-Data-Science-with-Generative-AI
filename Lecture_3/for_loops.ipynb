```python
#Basic Structure:
for i in range(1,11):
    print(i,"ssd")
```

```python
#Basic Structure:
for i in range(11,0,-1):
    print(i,"ssd")
```

```python
#Basic Structure:
for i in range(1,11,2):#where 2 is jumped/skip parameter.
    print(i,"ssd")
```

```python
for i in (1,2,3):
    print(i)
```
